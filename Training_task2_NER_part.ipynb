```python
import pandas as pd
import spacy
```

```python
df = pd.read_csv("cleaned_tokenized_tweets.csv")
texts = df["clean_text"].dropna().astype(str)
texts.head()
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def return_entities_and_processed_docs(text_series):
    named_entities = {}  # dictionary {ENTITY_TYPE: {ENTITY_TEXT: count}}
    processed_docs = []

    for item in text_series:
        doc = nlp(item)
        processed_docs.append(doc)

        for ent in doc.ents:
            entity_text = ent.text
            entity_type = ent.label_

            # focus on PERSON, ORG, GPE, DATE, TIME
            if entity_type in ["PERSON", "ORG", "GPE", "DATE", "TIME"]:
                # initialize sub-dictionary if not exists
                if entity_type not in named_entities:
                    named_entities[entity_type] = {}

                current_entities = named_entities[entity_type]
                current_entities[entity_text] = current_entities.get(entity_text, 0) + 1

    return named_entities, processed_docs
```

```python
named_entities, processed_docs = return_entities_and_processed_docs(texts)
```

```python
def print_top_10(named_entities):
    for entity_type in named_entities:
        print(f"\n{entity_type}")
        entities = named_entities[entity_type]
        sorted_keys = sorted(entities, key=entities.get, reverse=True)

        for name in sorted_keys[:10]:
            if entities[name] > 1:
                print(f"  {name}: {entities[name]}")
```

```python
print_top_10(named_entities)
```

```python
from spacy import displacy

# visualize the entities with its type in sentence
test = nlp("Mohammed visited Ali in Saudi Arabia on January 5th.")
displacy.render(test, style="ent", jupyter=True)

```

```python
def visualize_given_ent_type(processed_docs, entity, ent_type):
    for doc in processed_docs:
        for sentence in doc.sents:
            if entity in [ent.text for ent in sentence.ents if ent.label_ == ent_type]:
                displacy.render(sentence, style="ent", jupyter=True)

visualize_given_ent_type(processed_docs, "john cena", "PERSON")
```

```python
visualize_given_ent_type(processed_docs, "google", "ORG")
```
